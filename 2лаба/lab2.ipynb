```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# Генеруємо x від 0 до 10
x = np.linspace(0, 10, 50)

# Ідеальні значення y
y_true = 2 * x + 5

# Додаємо випадковий шум до y_true для отримання "виміряних" значень
np.random.seed(42)  # для відтворюваності
noise = np.random.normal(0, 2, size=x.shape)  # середнє 0, стандартне відхилення 2
y_measured = y_true + noise
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y_true, label='Ідеальна функція', color='blue')
plt.scatter(x, y_measured, label='Виміряні значення', color='red', alpha=0.7)
plt.title("Лінійна функція з похибками")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
mae = mean_absolute_error(y_true, y_measured)
mse = mean_squared_error(y_true, y_measured)

print(f"MAE (середня абсолютна похибка): {mae:.2f}")
print(f"MSE (середньоквадратична похибка): {mse:.2f}")
```

```python
df_results = pd.DataFrame({
    'x': x,
    'y_true': y_true,
    'y_measured': y_measured,
    'error': y_measured - y_true
})

# Додаємо рядок з метриками в кінець (опціонально)
df_metrics = pd.DataFrame({
    'x': [np.nan],
    'y_true': [np.nan],
    'y_measured': [np.nan],
    'error': [np.nan],
    'MAE': [mae],
    'MSE': [mse]
})

# Об’єднуємо таблиці
df_final = pd.concat([df_results, df_metrics], ignore_index=True)
```

```python
df_final.to_csv("results_with_errors.csv", index=False)
print("Результати збережено у файл results_with_errors.csv")
```
