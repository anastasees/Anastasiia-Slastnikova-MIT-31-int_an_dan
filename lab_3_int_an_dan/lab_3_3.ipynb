```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
```

```python
n_samples_n = 500
np.random.seed(0)

X1_n = np.random.uniform(10, 100, n_samples_n)
X2_n = np.random.normal(5, 2, n_samples_n)

X_numpy = np.stack([X1_n, X2_n], axis=1)

epsilon_n = np.random.normal(0, 5, n_samples_n)
y_numpy = 3.5 * X1_n - 1.2 * X2_n + 50 + epsilon_n

print("--- Кейс 3: Дані, Створені за допомогою NumPy ---")
print(f"Форма X_numpy (Ознаки): {X_numpy.shape}")
print(f"Форма y_numpy (Цільова): {y_numpy.shape}")
```

```python
X_train_n, X_test_n, y_train_n, y_test_n = train_test_split(
    X_numpy, y_numpy, test_size=0.3, random_state=42
)

print(f"Дані розділено. Тренувальна вибірка: {X_train_n.shape}")
```

```python
model_numpy = LinearRegression()
model_numpy.fit(X_train_n, y_train_n)
y_pred_n = model_numpy.predict(X_test_n)

print("Модель навчена та прогнози зроблені.")
```

```python
mse_n = mean_squared_error(y_test_n, y_pred_n)
r2_n = r2_score(y_test_n, y_pred_n)

print("Результати Регресійного Аналізу (Кейс 3: NumPy)")
print(f"Коефіцієнт детермінації (R²): {r2_n:.4f}")
print(f"Середньоквадратична помилка (MSE): {mse_n:.2f}")

print("\nКоефіцієнти моделі:")
for i, coef in enumerate(model_numpy.coef_):
    print(f"Ознака X{i+1}: {coef:.4f}")
print(f"Перетин (Intercept, b₀): {model_numpy.intercept_:.4f}")
```
