```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np 
```

```python
FILE_NAME = 'train.csv'
TARGET_COLUMN = 'y'
FEATURE_COLUMN = 'x'

try:
    df_kaggle = pd.read_csv(FILE_NAME)
    print(f"--- Кейс 2: Набір Даних з Kaggle ({FILE_NAME}) ---")
    print("Перші 5 рядків даних:")
    df_kaggle.head()
except FileNotFoundError:
    print(f"ПОМИЛКА: Файл '{FILE_NAME}' не знайдено.")
    df_kaggle = None
except KeyError:
    print(f"ПОМИЛКА: Не знайдено колонки '{TARGET_COLUMN}' або '{FEATURE_COLUMN}'.")
    df_kaggle = None
```

```python
if df_kaggle is not None:
    rows_before = len(df_kaggle)
    df_kaggle.dropna(inplace=True)
    rows_after = len(df_kaggle)
    
    if rows_before != rows_after:
        print(f"Видалено {rows_before - rows_after} рядків із пропущеними значеннями.")
    else:
        print("Пропущені значення відсутні або вже оброблені.")

    X_kaggle = df_kaggle[[FEATURE_COLUMN]].values
    y_kaggle = df_kaggle[TARGET_COLUMN].values
    
    X_train_k, X_test_k, y_train_k, y_test_k = train_test_split(
        X_kaggle, y_kaggle, test_size=0.3, random_state=42
    )
    
    print(f"Дані розділено. Тренувальна вибірка: {X_train_k.shape}")
```

```python
if df_kaggle is not None:
    model_kaggle = LinearRegression()
    model_kaggle.fit(X_train_k, y_train_k)
    y_pred_k = model_kaggle.predict(X_test_k)
    
    print("Модель навчена та прогнози зроблені.")
```

```python
if df_kaggle is not None:
    mse_k = mean_squared_error(y_test_k, y_pred_k)
    r2_k = r2_score(y_test_k, y_pred_k)

    print("--- Результати Регресійного Аналізу (Кейс 2: Синтетичні дані з Kaggle) ---")
    print(f"Коефіцієнт детермінації (R²): {r2_k:.4f}")
    print(f"Середньоквадратична помилка (MSE): {mse_k:.2f}")
    
    print(f"\nКоефіцієнт моделі (вага ознаки '{FEATURE_COLUMN}'): {model_kaggle.coef_[0]:.4f}")
    print(f"Перетин (Intercept, b₀): {model_kaggle.intercept_:.4f}")
```
