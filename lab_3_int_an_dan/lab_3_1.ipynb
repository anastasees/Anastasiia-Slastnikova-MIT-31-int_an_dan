```python
import numpy as np
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
n_samples = 300
n_features = 4
noise_level = 10.0 # –†—ñ–≤–µ–Ω—å —à—É–º—É –¥–ª—è —Ä–µ–∞–ª—ñ—Å—Ç–∏—á–Ω–æ—Å—Ç—ñ

X_gen, y_gen = make_regression(
    n_samples=n_samples,
    n_features=n_features,
    n_informative=3,     # –¢—ñ–ª—å–∫–∏ 3 –æ–∑–Ω–∞–∫–∏ —Ä–µ–∞–ª—å–Ω–æ –≤–ø–ª–∏–≤–∞—é—Ç—å –Ω–∞ 'y'
    n_targets=1,
    noise=noise_level,
    random_state=42
)

print(f"‚úÖ –î–∞–Ω—ñ –∑–≥–µ–Ω–µ—Ä–æ–≤–∞–Ω–æ.")
print(f"–§–æ—Ä–º–∞ X (–û–∑–Ω–∞–∫–∏): {X_gen.shape}")
print(f"–§–æ—Ä–º–∞ y (–¶—ñ–ª—å–æ–≤–∞ –∑–º—ñ–Ω–Ω–∞): {y_gen.shape}")
```

```python
X_train_gen, X_test_gen, y_train_gen, y_test_gen = train_test_split(
    X_gen, 
    y_gen, 
    test_size=0.3, # 30% –¥–∞–Ω–∏—Ö –ø—ñ–¥–µ –Ω–∞ —Ç–µ—Å—Ç—É–≤–∞–Ω–Ω—è
    random_state=42
)

print(f"\n‚úÖ –î–∞–Ω—ñ —Ä–æ–∑–¥—ñ–ª–µ–Ω–æ.")
print(f"–ù–∞–≤—á–∞–ª—å–Ω–∞ –≤–∏–±—ñ—Ä–∫–∞: {X_train_gen.shape[0]} –∑—Ä–∞–∑–∫—ñ–≤")
print(f"–¢–µ—Å—Ç–æ–≤–∞ –≤–∏–±—ñ—Ä–∫–∞: {X_test_gen.shape[0]} –∑—Ä–∞–∑–∫—ñ–≤")
```

```python
model_gen = LinearRegression()
model_gen.fit(X_train_gen, y_train_gen)

# –ü—Ä–æ–≥–Ω–æ–∑—É–≤–∞–Ω–Ω—è –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–º—É –Ω–∞–±–æ—Ä—ñ
y_pred_gen = model_gen.predict(X_test_gen)
```

```python
mse_gen = mean_squared_error(y_test_gen, y_pred_gen)
r2_gen = r2_score(y_test_gen, y_pred_gen)

print("\n--- üìä –†–µ–∑—É–ª—å—Ç–∞—Ç–∏ –†–µ–≥—Ä–µ—Å—ñ–π–Ω–æ–≥–æ –ê–Ω–∞–ª—ñ–∑—É (–ö–µ–π—Å 1: –ó–≥–µ–Ω–µ—Ä–æ–≤–∞–Ω—ñ –¥–∞–Ω—ñ) ---")
print(f"–ö–æ–µ—Ñ—ñ—Ü—ñ—î–Ω—Ç –¥–µ—Ç–µ—Ä–º—ñ–Ω–∞—Ü—ñ—ó (R¬≤): {r2_gen:.4f}")
print(f"–°–µ—Ä–µ–¥–Ω—å–æ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–∞ –ø–æ–º–∏–ª–∫–∞ (MSE): {mse_gen:.2f}")
```

```python
print("\n–ö–æ–µ—Ñ—ñ—Ü—ñ—î–Ω—Ç–∏ –º–æ–¥–µ–ª—ñ (–≤–∞–≥–∏ –æ–∑–Ω–∞–∫):")
for i, coef in enumerate(model_gen.coef_):
    print(f"  –û–∑–Ω–∞–∫–∞ X{i+1}: {coef:.2f}")
print(f"–ü–µ—Ä–µ—Ç–∏–Ω (Intercept, b‚ÇÄ): {model_gen.intercept_:.2f}")
```
